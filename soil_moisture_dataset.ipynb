```python
import pandas as pd

df = pd.read_csv("cleaned_data.csv")

total = len(df)
print("Total Data:", total)
```

```python
missing = df.isnull().sum().sum()

print("Missing Values:", missing)
```

```python
accuracy = 1 - (missing / total)

print("Accuracy:", accuracy)
```

```python
non_null = df.notnull().sum().sum()
total_cells = df.size

completeness = non_null / total_cells

print("Completeness:", completeness)
```

```python
df["date"] = pd.to_datetime(df[["year","month","day"]])
```

```python
latest_date = df["date"].max()

timeliness_data = df[df["date"] >= latest_date - pd.Timedelta(days=30)]

timeliness = len(timeliness_data) / len(df)

print("Timeliness:", timeliness)
```

```python
print("=== DATA QUALITY SCORE ===")
print("Accuracy:", round(accuracy,2))
print("Completeness:", round(completeness,2))
print("Timeliness:", round(timeliness,2))
```
